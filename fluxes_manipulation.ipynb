```python
import numpy
from numpy import *
import csv
import pandas as pd 
```

```python
data_dir = os.path.abspath("data")

path = os.path.join(data_dir,"Fluxes_Data.csv")
df = pd.read_csv(path) 
df
```

```python
df.loc[1,"rate"]
```

```python
list_1 = []
for r in df.index:
    number_ = df.loc[r,"rate"].split("±")
    # print(number_)
    df.loc[r,"rate"] = number_[0]
    add_on = number_[1]
    list_1.append(add_on)
df["SD_1"]= list_1
df
```

```python
list_1 = []
for r in df.index:
    number_ = df.loc[r,"rate_1"].split("±")
    # print(number_)
    df.loc[r,"rate_1"] = number_[0]
    add_on = number_[1]
    list_1.append(add_on)
df["SD_2"]= list_1
df
```

```python
df.to_csv(path)
```
